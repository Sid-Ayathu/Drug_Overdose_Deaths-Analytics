```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('clean8.csv')
```

```python

fig = px.violin(df, 
                y="num_of_deaths_in_county",  # Replace with the actual column for deaths
                x="Heroin",  # Replace with the actual column for Heroin
                box=True,  # Show box plot inside the violin plot
                points=False,  # Display individual points
                color="Heroin",  # Optional: color by Heroin
                title="Violin Plot of Number of Deaths vs Heroin for All Counties")

fig.update_layout(
    yaxis_title="Number of Deaths",
    xaxis_title="Heroin (Y/N)",
    height=1000  # Adjust the height to elongate the plot
)


fig.show()
```

```python
# Filter data for Hartford
hartford_data = df[df['DeathCounty'] == 'HARTFORD']

# Create the violin plot
fig = px.violin(hartford_data,
                y="Age",  # Dependent variable
                x="num_of_deaths_in_county",  # Independent variable
                box=True,  # Include box plot inside the violin
                points=False,  # Display individual data points
                title="Distribution of Number of Deaths in Hartford Across Age",
                color_discrete_sequence=["orange"])  # Customize color

# Update layout
fig.update_layout(
    xaxis_title="Number of Deaths",
    yaxis_title="Age",
    showlegend=False,
    height=1400
)

# Show the plot
fig.show()
```
