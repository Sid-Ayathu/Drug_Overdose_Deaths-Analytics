```python
import numpy as np
import pandas as pd
```

## __COMBINING Original A1 dataset and supplement A3 dataset into__ 

```python
df = pd.read_csv('datasets\starting datasets\drug_deaths.csv')
sup_df = pd.read_csv('datasets\starting datasets\drug_deaths_supplement(A3).csv')
```

```python
sup_df
```

```python
# List of columns to process
columns_to_process = [
    'Heroin', 'Cocaine', 'Fentanyl', 'Fentanyl_Analogue', 'Oxycodone',
    'Oxymorphone', 'Ethanol', 'Hydrocodone', 'Benzodiazepine', 'Methadone',
    'Amphet', 'Tramad', 'Hydromorphone', 
    'OpiateNOS', 'AnyOpioid'
]

# Convert 'Y' to 1 and other values to 0
sup_df[columns_to_process] = sup_df[columns_to_process].applymap(lambda x: 1 if x == 'Y' else 0)
```

```python

sup_df['no_of_drugs'] = sup_df[columns_to_process].sum(axis=1)
############################################################################################################
sup_df['InjuriesPerRace'] = sup_df.groupby('Race')['Race'].transform('count')
############################################################################################################
# Group by 'death county' and count the number of deaths in each county
sup_df['DeathsPercounty'] = sup_df.groupby('DeathCounty')['DeathCounty'].transform('count')
############################################################################################################
sup_df['ResidentsPerCounty'] = sup_df.groupby('ResidenceCity')['ResidenceCity'].transform('count')
############################################################################################################
# Group by 'DeathCity' and count the number of deaths per city
sup_df['InjuriesPerCounty'] = sup_df.groupby('InjuryCounty')['InjuryCounty'].transform('count')
############################################################################################################
# Add a new column with the count of residents per city
sup_df['ResidentsPerCity'] = sup_df.groupby('ResidenceCity')['ResidenceCity'].transform('count')
############################################################################################################
sup_df['InjuriesPerCity'] = sup_df.groupby('InjuryCity')['InjuryCity'].transform('count')
############################################################################################################
sup_df['DeathsPerCity'] = sup_df.groupby('DeathCity')['DeathCity'].transform('count')

sup_df
```

```python
df['Fentanyl'] = pd.to_numeric(df['Fentanyl'], errors='coerce').astype('Int64')
df['AnyOpioid'] = pd.to_numeric(df['AnyOpioid'], errors='coerce').astype('Int64')
```

```python
df['no_of_drugs'] = df[columns_to_process].sum(axis=1)
############################################################################################################
df['InjuriesPerRace'] = df.groupby('Race')['Race'].transform('count')
############################################################################################################
# Group by 'death county' and count the number of deaths in each county
df['DeathsPercounty'] = df.groupby('DeathCounty')['DeathCounty'].transform('count')
############################################################################################################
df['ResidentsPerCounty'] = df.groupby('ResidenceCity')['ResidenceCity'].transform('count')
############################################################################################################
# Group by 'DeathCity' and count the number of deaths per city
df['InjuriesPerCounty'] = df.groupby('InjuryCounty')['InjuryCounty'].transform('count')
############################################################################################################
# Add a new column with the count of residents per city
df['ResidentsPerCity'] = df.groupby('ResidenceCity')['ResidenceCity'].transform('count')
############################################################################################################
df['InjuriesPerCity'] = df.groupby('InjuryCity')['InjuryCity'].transform('count')
############################################################################################################
df['DeathsPerCity'] = df.groupby('DeathCity')['DeathCity'].transform('count')

df
```

```python
# Concatenate the DataFrames while preserving all columns
combined_df = pd.concat([df,sup_df], ignore_index=True)


# Save the combined DataFrame to a new CSV
combined_df.to_csv('combined2.csv', index=False)
combined_df
```
